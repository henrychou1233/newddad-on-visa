```python
from PIL import Image
import os

# 函数来将单个 JPG 转换为 PNG
def convert_jpg_to_png(jpg_file, png_file):
    try:
        # 打开 JPG 文件
        with Image.open(jpg_file) as img:
            # 将图像转换为 PNG 并保存
            img.save(png_file, "PNG")
        return f"转换成功: '{jpg_file}' 到 '{png_file}'"
    except Exception as e:
        return f"发生错误: {e}"

# 指定包含 JPG 文件的文件夹路径
jpg_folder = '/home/anywhere3090l/Desktop/henry/jjmvtec/VisA_dataset/macaroni2/test/bad'  # 请替换为包含 JPG 文件的文件夹路径

# 指定用于保存 PNG 文件的目标文件夹路径
png_folder = '/home/anywhere3090l/Desktop/henry/jjmvtec/VisA_dataset/macaroni2/test/bad'  # 请替换为用于保存 PNG 文件的目标文件夹路径

# 遍历文件夹中的所有 JPG 文件并进行转换
for root, _, files in os.walk(jpg_folder):
    for file in files:
        if file.lower().endswith(".jpg"):
            jpg_file_path = os.path.join(root, file)
            # 构建 PNG 文件的路径，替换文件扩展名为 .png
            png_file_path = os.path.join(png_folder, os.path.splitext(file)[0] + ".png")
            
            # 调用函数进行转换
            result = convert_jpg_to_png(jpg_file_path, png_file_path)
            print(result)  # 添加这一行来打印结果

```
